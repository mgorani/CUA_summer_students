```python
d = [1,2,3,4]
e = [5,6,7,8]

print(d)
print(e)
```

```python
import matplotlib.pyplot as plt

plt.scatter(d,e)
plt.title('d vs e', fontsize =20)
plt.xlabel('d')
plt.ylabel('e')
plt.show()
```

```python
import numpy as np

m,b = np.polyfit(d,e,1)

plt.scatter(d,e)

plt.plot(d, int(round(m))*d + b)
plt.title('d vs e', fontsize =20)
plt.xlabel('d')
plt.ylabel('e')
plt.show()
```
