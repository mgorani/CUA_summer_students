```python
import matplotlib.pyplot as plt
import numpy as np

W = np.array([1.94, 2.00, 2.14])
sig = np.array([170.0, 162.7, 152.1])
dsig = np.array([12.3, 11.9, 12.8])

print(W)
print(sig)
print(dsig)
```

```python
sig_max = np.add(sig,dsig)

print(sig_max)
```

```python
sig_min = np.subtract(sig, dsig)

print(sig_min)
```

```python
print(W)
print(sig)

plt.scatter(W,sig)
plt.title('W vs sig', fontsize =20)
plt.xlabel('W (Gv/c^2)')
plt.ylabel('sig')
plt.show()
```

```python
m,b = np.polyfit(W,sig,1)

print(W)
print(sig)

plt.scatter(W,sig)
plt.plot(W, int(round(m))*W + b)
plt.title('W vs sig', fontsize =20)
plt.xlabel('W (Gv/c^2)')
plt.ylabel('sig')
plt.show()
```

```python
j,k = np.polyfit(W,sig_min,1)

plt.scatter(W,sig_min)
plt.plot(W, int(round(j))*W + k)

a,c = np.polyfit(W,sig_max,1)

plt.scatter(W,sig_max)
plt.plot(W, int(round(a))*W + c)

plt.scatter(W,sig)
plt.plot(W, int(round(m))*W + b)
plt.title('W vs sig', fontsize =20)
plt.xlabel('W (Gv/c^2)')
plt.ylabel('sig')
plt.show()
```
