```python
# Fit Q2 dependence at W=2.2, t=0.5

import matplotlib.pyplot as plt
import numpy as np 

Q2 = [1.90, 2.35]
sig = [162.7,118.1]
dsig = [11.9, 9.0]

print (Q2)
print (sig)
print (dsig)
```

```python
pts = np.array([(1.90,2.35),(162.7, 118.1), (11.9, 9.0)])

print(pts)

f = plt.figure(figsize=(11.69,8.27))

Q2 = pts[0,:]
sig = pts[1,:]
dsig = pts[2,:]

print(Q2,"\n",sig)

plt.scatter(Q2,sig)

m, b = np.polyfit(Q2, sig, 1)

plt.plot(Q2, m*Q2 + b, label = 'Best Fit Line')
plt.legend()
plt.errorbar(Q2,sig,dsig, fmt='')
plt.title('$Q^2$ vs $\sigma$ at W=2.2, t=0.5', fontsize =20)
plt.xlabel('$Q^2$ (Ge$V^2$)')
plt.ylabel('$\sigma$ (nbar/$c^2$)')

plt.show()
```

```python
max_y = [174.6, 109.1]
min_y = [150.8,127.1]

print (max_y)
print (min_y)
```

```python
plt.scatter(Q2,sig)
plt.scatter(Q2,max_y)
plt.scatter(Q2,min_y)

p,z = np.polyfit(Q2, max_y, 1)
plt.plot(Q2, p*Q2 + z, label = 'Max Slope')

s,w = np.polyfit(Q2, min_y, 1)
plt.plot(Q2, s*Q2 + w, label = 'Min Slope')

m, b = np.polyfit(Q2, sig, 1)

plt.plot(Q2, m*Q2 + b, label = 'Best Fit Line')
plt.legend()
plt.errorbar(Q2,sig,dsig, fmt='')
plt.title('$Q^2$ vs $\sigma$ at W=2.2, t=0.5', fontsize =20)
plt.xlabel('$Q^2$ (Ge$V^2$)')
plt.ylabel('$\sigma$ (nbar/$c^2$)')



plt.show()
```
