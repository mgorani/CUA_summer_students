```python
import matplotlib.pyplot as plt
import numpy as np

pts = np.array([(0.19, 0.21, 0.243, 0.325, 0.417, 0.414), (900, 1300, 380, 330, 290, 150), (200, 100, 10, 10, 10, 10)])

print(pts)

f = plt.figure(figsize=(11.69,8.27))

x_1 = pts[0,:]
y_1 = pts[1,:]
y_unc = pts[2,:]

print(x_1,"\n",y_1)

plt.scatter(x_1,y_1)
m, b = np.polyfit(x_1, y_1, 1)

plt.plot(x_1, m*x_1 + b, label = 'Best Fit Line')
plt.legend()
plt.errorbar(x_1,y_1,y_unc, fmt = ' ')
plt.title('t vs $\sigma$ at $Q^2$=1.6, W=2.2, & x=0.1-0.2', fontsize =20)
plt.xlabel('t (Ge$V^2$)')
plt.ylabel(r'$\sigma$ ($\frac{nbar}{c^2}$)')

plt.show()
```

```python
correlation_matrix = np.corrcoef(x_1, y_1)
correlation_xy = correlation_matrix[0,1]
r_squared = correlation_xy**2

print (r_squared)
```
